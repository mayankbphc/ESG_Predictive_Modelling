```python
ESG_Data = pd.read_csv("ESG_Data.csv")
ESG_Data.head()
```

```python
ESG_Data.shape
```

```python
ESG_Data.dtypes
```

```python
ESG_Data = ESG_Data.iloc[:,:65]
ESG_Data.head()
```

```python
ESG_Data = ESG_Data.melt(id_vars = ["Country Name", "Country Code", "Indicator Name", "Indicator Code"],
             var_name = "Years",
             value_name = "Value")
```

```python
ESG_Data['Years'] = pd.to_numeric(ESG_Data['Years'])
ESG_Data
```

```python
ESG_theme_mapping = pd.read_excel("ESG Theme Mapping.xlsx")
ESG_theme_mapping.head()
```

```python
ESG_Data = ESG_Data.merge(ESG_theme_mapping, on='Indicator Name', how='left')
ESG_Data.head()
```

```python
ESG_Data.shape
```

```python
Final_data_by_Emissions_Data = pd.merge(left=Emission_Data, right=ESG_Data, how='left', left_on=['Code', 'Year'], right_on=['Country Code', 'Years'])
Final_data_by_Emissions_Data.head()
```

```python
Final_data_by_Emissions_Data.shape
```

```python
Final_data_by_Emissions_Data.drop(["Country Name", "Country Code", "Years"], axis = 1, inplace = True)
Final_data_by_Emissions_Data.head()
```

```python
Final_data_by_Emissions_Data.shape
```

```python
Final_data_by_ESG_Data = pd.merge(left=Emission_Data, right=ESG_Data, how='right', left_on=['Code', 'Year'], right_on=['Country Code', 'Years'])
Final_data_by_ESG_Data.head()
```

```python
Final_data_by_ESG_Data.drop(["Entity", "Code", "Year"], axis = 1, inplace = True)
Final_data_by_ESG_Data.head()
```

```python
Final_data_by_ESG_Data.shape
```

```python
Final_data_by_Emissions_Data.to_csv("Final_data_by_Emissions_Data.csv")
```
